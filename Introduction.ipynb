### What is known about the object?

Its official title is: *Illuminations of Vellumm*

It is a book the dates back to the early 1800s and was originally created by Esther Cory.

The book is filled with appliques that were extracted by Cory from a 14th Century carmelite missal.

The book is currently housed in the Archive and Special Collections of the University of Glasgow library.

<img src="Full_Alphabet.jpg" width="250" height="250">

As shown above the appliques include cut-outs of the letters from the missal.
