# Part 1 - Project Explanation

---

### What is the project?

Inspired by the original context of the creation of the text this project is:

**AI Powered Medieval Font Style**

This project will take images taken of the different applique lettering and train a generative AI model to recreate a similar style of lettering that anyone can use.

---

### What is the reasoning behind this project?

The main themes of this project are:
- Reinterpretation
- Influence

Overall, it focuses on the cycle of art i.e. its reproduction over time.

Benjamin (2018) discusses the concept of transformation and reproduction in the art sphere, how it is endlessley reproduce and recontextualised over time. He argues that reproduction changes meaning rather tha simply diluting originality.

Similarly, Bloom (1973), identified the anxiety that lies in artistic creation and highlights the struggle between artists who came before. Overall, arguing that art emerges by reworking, or revising older pieces.

This is to say that the concept of the cycle of art replication is well documented. Older art exists through the creation of new art, and at every level influences modern artists in their creation.

This is shown in the story of 'Illuminations of Vellumm', Cory took inspiration from the original cultural item of the missal and formed her own art work from that.

This project's main aim is to continue the cycle of art, highlighting the importance that cultural objects, such as 'Illuminations of Vellumm' or the original carmelite missal, have in the ongoing formation of modern art.

This will be done by following a similar path of inspiration that Cory took; she was inspired by the missal's lettering to creare her art, and this project is inspired by her use of lettering to form art.

In this way, the audience for the project is quite broad; the project is more of a general catalyst for thought and reflection. If one audience was to be chosen, art students who want to get thinking about how inspiration within art happens.

---

### How does the project allow for AI to aid the human creative process?

As discussed, the cycle of art is the key message of this project. With this in mind, the AI tool plays an assisting factor in connecting modern artists to those of the past, rather than it being the creator of the art.

In this way, the AI tool acts as the applique method that Cory made use of. It is the vessel for which the art will be created, acting as a bridge between Cory's and the art students' work.

Through users' use of the font style it reminds them of the original book, and empowers them to become inspired by Cory's art to create their own. The basis of which is formed by the AI generation of the font style.

---

### What are the goals for the project and target audience (art students)?

1. Using pictures of the original object, train a generative AI model to replicate the applique lettering and create a font style based on this.
2. Through doing this, create an artistic tool that users can make their own medieval style writings/artwork with.
3. Make users think about where inspiration comes from, and how older art inspires their own art.
