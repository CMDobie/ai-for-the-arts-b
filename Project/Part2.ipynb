### What AI model will be used for this?

Initially, I was going to make the model myself and train the model with photos I have taken using the knowledge gained from part A of the course. However, after peer feedback and suggestions I decided it would be easier, and more likely to work if I found an existing model that has existing tools that I need for this project.

With this in mind, the search began to find the right model with the right toolset. For instance, certain generative AI models provide font generation (BasedLabs, Canvus). However, these models only had text-to-font allowance, whereas this project must be built using image-to-font. If only a text description is given of the archival object, then it will not be as connected to the theme of the life cycle of art. This is because it does not directly include the object.

Leading from this, the final AI tool was discovered and chosen to be used for this project: **GLIPH**.

---

### What is GLIPH?

The main functions of GLIPH include:

- AI Font generation
- AI-powered bitmap images to vector glyph conversion
- Allowance for user correction and manipulation to finesse details

Firstly, GLIPH meets the base requirement for what is needed: Font generation. However, it additionally allows for a range of in-depth options and analysis where other models would not.

Where other AI models would simply take the input and spit out the results without any explanation, GLIPH gives a detailed overview of the creation process and allows the user to make ammendments along the way so the finished product is what is wanted.

Initially, GLIPH works the same as a regular generative AI, input the material that will be analysed.
