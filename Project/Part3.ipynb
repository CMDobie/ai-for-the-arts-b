### How did the final product turn out?

The finished project works as intended and is fully functional. As mentioned previously, there are a few missing letters but this is through no fault of the AI tools used but the original object not including these in the first place.

---

### How to use and access the project?

The project can be downloaded from with the 'Project' folder, the file name is MedievalFontStyle.tff . Once it is downloaded onto the device, open the file and click the install button.

Note that the font will only be usable on device-side software including MS Word, and not server-side such as GitHub notebooks or Google Docs etc.
